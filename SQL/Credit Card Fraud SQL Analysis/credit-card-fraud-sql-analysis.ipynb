```python
import duckdb
```

```python
duckdb.sql("""
SELECT *
FROM fraud
LIMIT 5;
""").df()
```

```python
duckdb.sql("""
SELECT
    COUNT(*) AS Total_Transactions
FROM fraud;
""").df()
```

```python
duckdb.sql("""
SELECT
    Class,
    COUNT(*) AS Total
FROM fraud
GROUP BY Class;
""").df()
```

```python
duckdb.sql ("""
SELECT 
ROUND (100.0 * SUM(Class)/COUNT(*),4) AS Fraud_Percentage
FROM fraud;
""").df()
```

```python
duckdb.sql("""
SELECT
    ROUND(AVG(Amount),2) AS Average_Amount
FROM fraud;
""").df()
```

```python
duckdb.sql("""
SELECT 
  MAX (AMOUNT) AS Highest_Transaction
  FROM fraud;
  """).df()
```

```python
duckdb.sql("""
SELECT
Amount, Class
FROM fraud
ORDER BY Amount DESC
LIMIT 10;
""").df()
```

```python
duckdb.sql("""
SELECT
    ROUND(AVG(Amount),2) AS Average_Fraud_Amount
FROM fraud
WHERE Class = 1;
""").df()
```

```python
duckdb.sql("""
SELECT
    ROUND(AVG(Amount),2) AS Average_Genuine_Amount
FROM fraud
WHERE Class = 0;
""").df()
```

```python
duckdb.sql("""
SELECT
    Amount,
    Class
FROM fraud
WHERE Amount >
(
    SELECT AVG(Amount)
    FROM fraud
)
ORDER BY Amount DESC
LIMIT 20;
""").df()
```

```python
duckdb.sql("""
SELECT
    COUNT(*) AS High_Value_Fraud
FROM fraud
WHERE Class = 1
AND Amount > 1000;
""").df()
```

```python
duckdb.sql("""
SELECT *
FROM fraud
WHERE Class = 1
LIMIT 10;
""").df()
```

```python
duckdb.sql("""
SELECT *
FROM fraud
WHERE Class = 0
LIMIT 10;
""").df()
```

```python
duckdb.sql("""
SELECT
    Amount,
    Class
FROM fraud
WHERE Amount > 500
ORDER BY Amount DESC;
""").df()
```

```python
duckdb.sql("""
SELECT
    Amount,
    Class
FROM fraud
WHERE Amount < 10;
""").df()
```

```python
duckdb.sql("""
SELECT
    COUNT(*) AS Total_Transactions
FROM fraud
WHERE Amount > 1000;
""").df()
```

```python
duckdb.sql("""
SELECT
    ROUND(AVG(Amount),2) AS Average_Amount
FROM fraud
WHERE Amount > 100;
""").df()
```

```python
duckdb.sql("""
SELECT
    Amount,
    Class
FROM fraud
ORDER BY Amount ASC
LIMIT 10;
""").df()
```

```python
duckdb.sql("""
SELECT
    CASE
        WHEN Amount < 100 THEN 'Low'
        ELSE 'High'
    END AS Category,
    COUNT(*) AS Total_Transactions
FROM fraud
GROUP BY Category;
""").df()
```
